```python
# # Library importation

# These libraries are imported for data manipulation (pandas), numerical calculations (numpy), and visualization (matplotlib, seaborn).
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



# # Load the dataset

# Reads the CSV file into a pandas DataFrame for analysis.
file_path = 'C:/Users/DELL 5520/Desktop/My Data Portfolio Projects/Sales_Analysis/supermarket_sales.csv'
df = pd.read_csv(file_path)

# # Data Preprocessing

# Converts the Date column to a datetime object.
# Extracts Month, Day, and Hour for time-based analysis.
# Ensures Time is parsed correctly for extracting hours.
# Why? This prepares the data for trend and time-based analysis.

def preprocess_data(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df['Month'] = df['Date'].dt.to_period('M')
    df['Day'] = df['Date'].dt.day
    df['Hour'] = pd.to_datetime(df['Time']).dt.hour
    return df

df = preprocess_data(df)

# # Basic Analysis

# Calculates key metrics: total revenue, total quantity sold, average rating, and customer type distribution.
# Insights Derived:

# If one customer type (e.g., "Members") dominates, marketing efforts can focus on that group.
# Ratings help understand overall satisfaction.

def basic_analysis(df):
    total_revenue = df['Total'].sum()
    total_quantity = df['Quantity'].sum()
    average_rating = df['Rating'].mean()
    customer_type_distribution = df['Customer type'].value_counts(normalize=True) * 100

    print(f"Total Revenue: ${total_revenue:,.2f}")
    print(f"Total Quantity Sold: {total_quantity}")
    print(f"Average Customer Rating: {average_rating:.2f}")
    print("Customer Type Distribution (%):")
    print(customer_type_distribution)

basic_analysis(df)


# # Trend Analysis

# Aggregates revenue and quantity sold by month.
# Visualizes monthly trends to identify seasonality or growth patterns.
# Insights Derived:

# Identifies peak sales months to time promotions effectively.

def trend_analysis(df):
    # Ensure 'Month' is converted to a datetime or string format
    df['Month'] = df['Month'].dt.to_timestamp()  # Convert Period to Timestamp
    monthly_trend = df.groupby('Month').agg({'Total': 'sum', 'Quantity': 'sum'}).reset_index()

    plt.figure(figsize=(12, 6))
    sns.lineplot(data=monthly_trend, x='Month', y='Total', label='Revenue')
    sns.lineplot(data=monthly_trend, x='Month', y='Quantity', label='Quantity Sold')
    plt.title('Monthly Sales and Revenue Trends')
    plt.xlabel('Month')
    plt.ylabel('Values')
    plt.legend()
    plt.show()
    
trend_analysis(df)

# # Product Performance

# Analyzes sales and revenue per product line.
# Visualizes product-line revenue contributions.
# Insights Derived:

# Highlights top-performing product lines for prioritization.

def product_analysis(df):
    product_performance = df.groupby('Product line').agg({'Quantity': 'sum', 'Total': 'sum', 'Unit price': 'mean'}).sort_values(by='Total', ascending=False).reset_index()

    print("Top Performing Product Lines:")
    print(product_performance)

    plt.figure(figsize=(10, 5))
    sns.barplot(data=product_performance, x='Product line', y='Total')
    plt.title('Revenue by Product Line')
    plt.xticks(rotation=45)
    plt.show()

product_analysis(df)

# # Branch and City Performance

# Evaluates performance by branch and city.
# Provides insights on geographic trends.
# Insights Derived:

# Identifies underperforming locations for targeted interventions.

def regional_analysis(df):
    branch_performance = df.groupby('Branch').agg({'Quantity': 'sum', 'Total': 'sum'}).reset_index()
    city_performance = df.groupby('City').agg({'Quantity': 'sum', 'Total': 'sum'}).reset_index()

    print("Branch Performance:")
    print(branch_performance)

    print("City Performance:")
    print(city_performance)

    plt.figure(figsize=(8, 4))
    sns.barplot(data=branch_performance, x='Branch', y='Total')
    plt.title('Revenue by Branch')
    plt.show()

    plt.figure(figsize=(8, 4))
    sns.barplot(data=city_performance, x='City', y='Total')
    plt.title('Revenue by City')
    plt.show()

regional_analysis(df)

# # Customer Insights

# Analyzes sales by gender and customer type.
# Offers demographic insights for tailored marketing.

def customer_insights(df):
    gender_analysis = df.groupby('Gender').agg({'Quantity': 'sum', 'Total': 'sum'}).reset_index()
    customer_type_analysis = df.groupby('Customer type').agg({'Quantity': 'sum', 'Total': 'sum'}).reset_index()

    print("Gender-based Analysis:")
    print(gender_analysis)

    print("Customer Type Analysis:")
    print(customer_type_analysis)

    plt.figure(figsize=(6, 4))
    sns.barplot(data=gender_analysis, x='Gender', y='Total')
    plt.title('Revenue by Gender')
    plt.show()

    plt.figure(figsize=(6, 4))
    sns.barplot(data=customer_type_analysis, x='Customer type', y='Total')
    plt.title('Revenue by Customer Type')
    plt.show()

customer_insights(df)

# # Payment Methods Analysis

# Shows the proportion of different payment methods.
# Identifies customer preferences in payment.
# Insights Derived:

# Streamlines operations based on preferred payment methods.

def payment_analysis(df):
    payment_analysis = df['Payment'].value_counts(normalize=True) * 100

    print("Payment Method Distribution (%):")
    print(payment_analysis)

    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x='Payment', order=df['Payment'].value_counts().index)
    plt.title('Payment Methods Distribution')
    plt.show()

payment_analysis(df)

# Actionable Insights
print("\nActionable Insights:")
print("1. Focus marketing efforts on the most profitable product lines.")
print("2. Investigate strategies to boost sales in low-performing branches and cities.")
print("3. Leverage customer segmentation to create targeted promotions.")
print("4. Optimize inventory for high-demand products based on trends.")
print("5. Encourage the use of payment methods that streamline operations.")
```
